Exploratory Data Analysis 



```python
# Data visualization through pandas

import pandas as pd
```

```python
data = {
    "Name" : ['Rishi','Sagar','Priya','Megha','Deep','Mihika'],
    "Scores" : [80,90,85,77,91,79]    
}
```

```python
students = ['Rishi','Sagar','Priya','Megha','Deep','Mihika']
df = pd.DataFrame(data,index = students)
```

```python
df
```

```python
# line plot

df.plot()
```

```python
df.plot(kind='bar')
```

```python
df.plot(kind = 'barh')
```

```python
df.plot(kind = 'hist')
```

```python
df.plot(kind = 'kde')
```

```python
df.plot(kind = 'area')
```

```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df1 = sns.load_dataset('iris')
df1
```

```python
df1.value_counts("species")
```

```python
plot = df1.plot(figsize=(10,5),title = 'IRIS DATASET')
plot.set_xlabel("X - Axis")
plot.set_ylabel("Y - Axis")
```

```python
df1 = df1.drop(['species'],axis = 1)
```

```python
df1.iloc[0:10]
```

```python
df1.iloc[0:10].plot(kind = 'bar')
```

```python
df1.plot.scatter(x = 'sepal_length',y = 'sepal_width')
```

```python
df1.iloc[0:20].plot.pie(y= 'sepal_length',subplots=True,figsize=(5,5))
```
