```python
import pandas as pd
```

```python
df = pd.read_csv(r"Advertising_data.csv")
```

```python
df.head()
```

```python
less than 10000 ---> not eligible for PL
10000 ----> 1 Lac (max 3 years)
20k ---> 2 Lac
1 Lac --> 10 Lac
```

```python
def eligibility(salary):
    result = salary/10000
    if result<1:
        return "Not eligible"

    else:
        rs = int(result)*1 
        return str(rs) + " Lac"
```

```python
df["Loan Eligibility"] = df["EstimatedSalary"].apply(eligibility)
```
