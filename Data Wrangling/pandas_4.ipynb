```python
import pandas as pd
```

```python
dict1 = {
    "Brand" : ['Maruti', 'Hyundai', 'Tata', 'Mahindra', 'Renault', 'Tata', 'Maruti','Hyundai', 'Tata', 'Mahindra', 'Renault', 'Tata'],
    "Year" : [2012,2013,2014,2008,2010,2011,2009,2012,2014,2015,2018,2016],
    "Kms Driven" : [50000,60000,25000,75000,49000,30000,23000,10000,50000,60000,25000,30000],
    "City" : ["Delhi", "Mumbai","Gurgaon","Mumbai","Hyderabad","Chennai","Bengaluru","Kolkata","Delhi", "Mumbai","Gurgaon","Mumbai"],
    "Mileage" : [30,20,25,28,34,19,20,29,26,15,35,22]
}
```

```python
df = pd.DataFrame(dict1)
```

```python
df
```

```python
df.info()
```

```python
df['Mileage'].mean()
```

```python
def Rating(mileage):
    if mileage >=25:
        return "Good"
    else:
        return "Poor"
```

```python
df["Mileage Rating"] = df["Mileage"].apply(Rating)
```

```python
df
```

```python
df['Country'] = "India"
```

```python
df
```

```python
def cartype(year, kms):
    if year>=2012 and kms <=30000:
        return "New Car"

    else:
        return "Old Car"
```

```python
df['Car Type'] = list(map(cartype,df['Year'],df['Kms Driven']))
```

```python
df
```

```python
# nlargest()

df.nlargest(3,"Mileage")
```

```python
# nsmallest()

df.nsmallest(1,"Mileage")
```
